#### Задание 1

В одном графическом окне создать 4 подобласти, в первой из них построить график функции $y=f(x)$ на промежутке $[-2;2]$, где $f(x)=-e^{-x}+2$, в остальных областях на том же промежутке построить графики функций: 
* $y_{1} = f(x+2)$
* $y_{2} = f(\left |x+2  \right |)$
* $y_{3} = \left |f(\left |x+2  \right |)  \right |+1$

**В отчёт:** вставить рисунок, описать своими словами последовательность преобразований графиков $y \rightarrow y_{1} \rightarrow y_{2} \rightarrow y_{3}$

```python
import numpy as np
import matplotlib.pyplot as plt

# задаем пределы для X
x = np.linspace(-2,2,1000)

# функция 
def f(x):
    return -np.exp(1)**(-x)+2

f_x = f(x)
f_y1 = f(x+2)
f_y2 = f(np.abs(x+2))
f_y3 = np.abs(f(np.abs(x+2))) + 1

def create_plot(area, a1,a2,a3, x, y, title, show_title = True):
    area.add_subplot(a1,a2,a3)
    area.set_dpi(50)
    font = {'size'   : 18}
    plt.rc('font', **font)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axis([min(x), max(x), -1, 3])
    plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0.2, hspace=0.2)
    ax = plt.gca()
    ax.set_xlabel("X", fontsize=22,labelpad=10)
    ax.set_ylabel(title, fontsize=22, labelpad=10)
    if show_title:
        ax.set_title("График" +str(a3), loc="right",  fontsize=18,  pad=20)
    plt.axhline(color='k', linewidth=2)
    plt.axvline(color='k', linewidth=2)
    plt.grid(True)

    plt.plot(x,y,'-g', linewidth=3)

area = plt.figure(figsize=(17,17))

create_plot(area, 2,2,1, x, f_x, "$f(x) =-e^{-x}+2$")
create_plot(area, 2,2,2, x, f_y1, "$y_{1} = f(x+2)$")
create_plot(area, 2,2,3, x, f_y2, "$y_{2} = f(|x+2|)$")
create_plot(area, 2,2,4, x, f_y3, "$y_{3} = |f(|x+2|)|+1$")

plt.show()
```

На первый взгляд между графиками 2 и 3 нет никакой разницы. Увеличим промежуток построения до [-3,3], чтобы ее увидеть.

```python
x = np.linspace(-3,3,1000)

f_x = f(x)
f_y1 = f(x+2)
f_y2 = f(np.abs(x+2))
f_y3 = np.abs(f(np.abs(x+2))) + 1

area = plt.figure(figsize=(17,17))
sb = create_plot(area, 2,2,1, x, f_y1, "$y_{1} = f(x+2)$", show_title=False)
create_plot(area, 2,2,2, x, f_y2, "$y_{2} = f(|x+2|)$", show_title=False)
```

Из рисунка выше наглядно выше, что преобразование $y_{1} \rightarrow y_{2}$ отзеркаливает график по оси X, так как мы берем значение по модулю, а, следовательно, значение функции не может быть отрицательным. 

* $y \rightarrow y_{1}$ - сжимает график по оси y
* $y_{1} \rightarrow y_{2}$ - отзеркаливает график по оси x
* $y_{2} \rightarrow y_{3}$ - сдвигает график по оси y (в нашем случае на единицу вверх, так как мы прибавляем к значению 1)

#### Задание 2

Вычислить письменно и в Python, сделать рисунки на комплексной плоскости

* $z_{1} = \sqrt{3}+i\sqrt{3}$, $z_{2} = -i$ - радиус векторами синего и красного цвета, соответственно;

* $z_{3} = z_{1} + z_{2}$ - знаком «+» зеленого цвета;

* $z_{4} = z_{1}z_{2}$ - знаком «*» синего цвета;

* $z_{5} = z_{1}/z_{2}$ – знаком «*» красного цвета;

* $z_{6} = \sqrt[3]{z_{2}}$   – знаками «^» красного цвета.

Проверить письменно правильность произведенных вычислений и отображений. Вставить скан.

```python
# https://habr.com/ru/companies/ruvds/articles/565382/

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy.abc import x

z1 = complex(np.sqrt(3),np.sqrt(3))
z2 = -1j
z3 = z1+z2
z4 = z1*z2
z5 = z1/z2
z6 = np.array([complex(item) for item in sp.solve(sp.Eq(x**3, z2), x)])

font = {'size'   : 8}
plt.rc('font', **font)
plt.ylabel('imaginary') 
plt.xlabel('real') 
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.grid()

plt.quiver(z1.real, z1.imag, color='b', label='$z_{1} = \sqrt{3}+i\sqrt{3}$', scale=7)
plt.quiver(z2.real, z2.imag, color='r', label="$z_{2} = -i$", scale=7)
plt.plot(z3.real, z3.imag, 'g+', markersize=7, label="$z_{3} = z_{1} + z_{2}$")
plt.plot(z4.real, z4.imag, 'b*', markersize=7, label="$z_{3} = z_{1}z_{2}$")
plt.plot(z5.real, z5.imag, 'r*', markersize=7, label="$z_{3} = z_{1} + z_{2}$")
plt.plot(np.real(z6), np.imag(z6), '^r', markersize=7, label="$z_{6} = \sqrt[3]{z_{2}}$")

plt.legend(loc='lower left')
plt.show() 
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from sympy.abc import x

n, w = 6, 64

y = sp.Eq(x**n, w)
all_roots =  np.array([complex(root) for root in sp.solve(y, x)])
real_roots = np.array([root.real for root in all_roots if root.imag == 0])
sp.pprint(all_roots)

# настраиваем внешний вид гарфика
plt.title('$\sqrt[6]{64}$ complex roots')
font = {'size': 10}
plt.rc('font', **font)
plt.axis('equal')
plt.grid()

# добавляем оси x,y
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

# добавляем окружность, на которой лежат все корни
t = np.linspace(0,2*np.pi,100)
plt.plot(sp.Abs(w)**(1/n)*np.cos(t),sp.Abs(w)**(1/n)*np.sin(t),':g')

# добавляем сами корни
plt.plot(np.real(all_roots), np.imag(all_roots), 'dr')
plt.show()
```

#### Задание 5

Для последовательности $x_{n}=\frac{4-7n}{2n+1}$:
1) Найти $a = \lim_{n \to \infty }x_{n}$ письменно и в Python, **дать определение предела по Коши**

2) Вычислить $n_{0}(0.01), n_{0}(0.001)$
3) Построить два рисунка, которые иллюстрируют определение предела и результаты пункта 2

В отчёт: вставить значение $a, n_{0}(0.01), n_{0}(0.001)$, рисунки.

**Смысл задания** показать понимание определения предела по Коши с помощью рисунка.

```python
import sympy as sp
from sympy.abc import n

x = (4-7*n)/(2*n-1)

a = sp.limit_seq(x, n)
sp.pprint(x)
print()
sp.pprint(a)

n_0_1 = sp.solve_univariate_inequality(sp.Abs(x-a)<0.01, n,relational=False)
n_0_2 = sp.solve_univariate_inequality(sp.Abs(x-a)<0.001, n,relational=False)
b = sp.solve_univariate_inequality(n>0, n,relational=False)

intr1 = sp.Intersection(n_0_1, b)
intr2 = sp.Intersection(n_0_2, b)

sp.pprint(intr1)
sp.pprint(intr2)
```

В отчёт: вставить значение $a, n_{0}(0.01), n_{0}(0.001)$, рисунки.

```python
import sympy as sp
from sympy.abc import n

x = (2*n**2 + 2*n + 1)/(n**2+n-1)
a = sp.limit_seq(x, n)
sp.pprint(x)
print()

n_0_1 = sp.solve_univariate_inequality(sp.Abs(x - a)<0.01, n,relational=False)
n_0_2 = sp.solve_univariate_inequality(sp.Abs(x - a)<0.001, n,relational=False)

b = sp.solve_univariate_inequality(n>0, n,relational=False)

intr1 = sp.Intersection(n_0_1, b)
intr2 = sp.Intersection(n_0_2, b)

sp.pprint(intr1)
sp.pprint(intr2)
```
